### Box and Violin plot in Seaborn

```python
df.columns
```

```python
# box plot
plt.figure(figsize=(20,7))
sns.boxplot(data=df)
```

```python
# x - horizontal box plot
plt.figure(figsize=(10,5))
sns.boxplot(x='tip',data=df)
plt.ylabel('Tip Box Plot')
```

```python
# y - horizontal box plot
sns.boxplot(y='tip',data=df,color='yellow')
plt.xlabel('tip($)')
```

```python
plt.figure(figsize=(20,7))

sns.boxplot(x='day',y='tip',data=df,hue='sex')
plt.xlabel('Day') # discrete
plt.ylabel('tip($)')
plt.title('Day Wise - Outlier Detection with Hue as sex',size =15)
```

```python
plt.figure(figsize=(20,7))
sns.boxplot(x='day',y='tip',data=df)
plt.xlabel('Day')
plt.ylabel('tip($)')
plt.title('Day Wise - Outlier Detection without Hue',size = 15)
```

```python
plt.figure(figsize=(20,7))

sns.boxplot(x = df['tip'], y = df['total_bill'])   # since both are continuous

plt.xlabel('Tip ($)')
plt.ylabel('Total Bill($)')
plt.title('Not Recommended - Continuous Data on x & y ',size = 15) # on x - Discrete
```

```python
plt.figure(figsize=(15,7))
sns.boxplot(x='time',y='tip',data=df,hue='smoker')
plt.xlabel('time')
plt.ylabel('tip($)')

plt.title('variation of tip when time is changed')
```

```python
# variation of tip when time is changed  - by violin Plot
plt.figure(figsize=(15,7))
sns.violinplot(x='time',y='tip',data=df,hue='smoker')
plt.xlabel('time')
plt.ylabel('tip($)')
```

## Bivariate → “two variables” data visualization

### Strip Plot

```python
# variation of tip 
sns.stripplot(y='tip',data=df)
```

```python
# variation of tip - day wise trend with hue (differensiate) - sex(male/female)
sns.stripplot(x='day',y='tip',data=df,jitter=True, hue='sex')
plt.ylabel('tip ($)')
plt.xlabel('day')
```

```python
# variation of tip - day wise trend with hue (differensiate) - sex(male/female)
sns.stripplot(x='day',y='tip',data=df,jitter=False, hue='sex')
plt.ylabel('tip ($)')
plt.xlabel('day')
```

```python
# analysis of tip wrt to time of food

sns.stripplot(x='time',y='tip',data=df,hue='sex')
plt.xlabel('time')
plt.ylabel('tip($)')
```

### Swarm Plot

```python
sns.swarmplot(y = 'tip', data=df,hue = 'sex')
plt.ylabel('tip($)')
```

### Joint Plot

```python
"scatter" | "reg" | "resid" | "kde"(kernel density estimation) | "hex" 
```

```python
sns.jointplot(data=df,x='total_bill',y='tip') # by default kind = scatter
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='reg')  # regression

# very much usefulin Simple Linear Regression -
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='resid')  # residual plot - residual vs x(predictors/explanotary)
# very much usefulin Multi Linear Regression Assumption Validation
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='kde')
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='hex')
```
